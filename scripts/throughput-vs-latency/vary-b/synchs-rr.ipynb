```python
# Build a ping percentile plot

import matplotlib.pyplot as plt
from numpy import percentile
from csv import reader
import math

ROOT="/run/media/haxolotl/Backup/GitRoot/libapollo-rust"
```

```python
shs_b100_w = []
shs_b100_lat = []
shs_b100_tput = []

shs_b400_w = []
shs_b400_lat = []
shs_b400_tput = []

shs_b800_w = []
shs_b800_lat = []
shs_b800_tput = []

shs_b2000_w = []
shs_b2000_lat = []
shs_b2000_tput = []


BlockSize=100
Protocol="synchs-rr"
with open(f"{ROOT}/data/Jan20/c5/vary-b/b{BlockSize}-{Protocol}-run-cleaned.csv") as f:
    rdr = reader(f)
    for row in rdr:
        shs_b100_w.append(float(row[0])/1000)
        shs_b100_tput.append(float(row[1])/1000)
        shs_b100_lat.append(float(row[2]))

BlockSize=400
with open(f"{ROOT}/data/Jan20/c5/vary-b/b{BlockSize}-{Protocol}-run-cleaned.csv") as f:
    rdr = reader(f)
    for row in rdr:
        shs_b400_w.append(float(row[0])/1000)
        shs_b400_tput.append(float(row[1])/1000)
        shs_b400_lat.append(float(row[2]))

BlockSize=800
with open(f"{ROOT}/data/Jan20/c5/vary-b/b{BlockSize}-{Protocol}-run-cleaned.csv") as f:
    rdr = reader(f)
    for row in rdr:
        shs_b800_w.append(float(row[0])/1000)
        shs_b800_tput.append(float(row[1])/1000)
        shs_b800_lat.append(float(row[2]))

BlockSize=2000
with open(f"{ROOT}/data/Jan20/c5/vary-b/b{BlockSize}-{Protocol}-run-cleaned.csv") as f:
    rdr = reader(f)
    for row in rdr:
        shs_b2000_w.append(float(row[0])/1000)
        shs_b2000_tput.append(float(row[1])/1000)
        shs_b2000_lat.append(float(row[2]))

```

```python
plt.rc('text',usetex=True)

# plt.rcParams["figure.figsize"] =(6.6, 4)
plt.rcParams.update({'font.size': 13})

plt.plot(shs_b100_tput,shs_b100_lat, label="Sync HotStuff - RR (b100)", linestyle='--', marker='+')

plt.plot(shs_b400_tput,shs_b400_lat, label="Sync HotStuff - RR (b400)", linestyle='--', marker='+')

plt.plot(shs_b800_tput,shs_b800_lat, label="Sync HotStuff - RR (b800)", linestyle='--', marker='+')

plt.plot(shs_b2000_tput,shs_b2000_lat, label="Sync HotStuff - RR (b2000)", linestyle='--', marker='+')

plt.legend()
plt.xlabel("Response Rate (in Kops/s)")
plt.ylabel("Latency (in ms)")
# plt.show()
# plt.xticks(log_points)
plt.savefig(f"{ROOT}/Plots/vary-b/RR-Tput-vs-latency-for-diff-b.pdf", bbox_inches="tight")
```
